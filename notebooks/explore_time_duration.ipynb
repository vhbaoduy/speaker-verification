```python
import utils
import pandas as pd
import os
import glob
import tqdm
import matplotlib.pyplot as plt
```

```python
PATH = 'data/loop1'
```

```python
files = glob.glob(PATH + '/' + str(0) + '/*/*.wav')
```

```python
scr, sr = utils.load_audio(files[0], 16000)
```

```python
plt.plot(scr)
```

```python
data = {}
for digit in os.listdir(PATH):
    data[digit] = []
    samples = glob.glob(PATH + '/' + str(digit) + '/*/*.wav')
    for sample in tqdm.tqdm(samples):
        src, sr = utils.load_audio(sample, 16000)
        time = len(src) / 16000
        data[digit].append(time)
```

```python
df = {
    'min':[],
    'max':[]
}
for digit in data:
    min_ = min(data[digit])
    max_ = max(data[digit])
    df['min'].append(min_)
    df['max'].append(max_)
    
```

```python
df = pd.DataFrame(df, index=[i for i in range(0,10)])
```
